```python
#here we can see so many outliers in this column so we can treat it with IQR method
q1 = df['floor_area_sqm'].quantile(0.25)
q3 = df['floor_area_sqm'].quantile(0.75)
iqr = q3 - q1
lower_limit=q1-(1.5*iqr)
upper_limit=q3+(1.5*iqr)
print(lower_limit)
print(upper_limit)
```

```python
#if the number is greater than upper limit the values is chaned into upper_limit 
df['floor_area_sqm'] = df['floor_area_sqm'].apply(lambda x: min(x,upper_limit))
```
